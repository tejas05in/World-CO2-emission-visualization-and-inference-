# Correlation between population size and CO2 emission

```python
data=df[['Population  (2016)','CO2 Emissions  (tons, 2016)']]
```

```python
data.corr()
```

```python
x=data['Population  (2016)']
y=data['CO2 Emissions  (tons, 2016)']
```

```python
from scipy.stats import pearsonr
```

```python
r,p=pearsonr(x,y)
```

```python
print(r,p)
```

## **Inference of correlation**

**The pearson correlation coeffecient for population vs CO2 emission stands at 0.825 or 82.50% which tells us that for every one unit increase in population the emission goes up by 82.50% and this observation is statistically highly significant**
